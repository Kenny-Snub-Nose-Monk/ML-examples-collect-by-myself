## 项目要求
### 使用 Pandas 或者 Spark 加载给定的数据集，选出北京相关的数据，完成数据分析和可视化。
Tips：下面列出一些可分析的点，其他分析角度和切入点，都可能获得额外的加分
- 单维度分析：
    - 小区、户型、面积、朝向、热度等单维度的分布分析和可视化展示
    - 查看 Pandas 与 Seaborn 部分，选择合适的统计方法和可视化图类型
    - 请大家一定带上数据分析说明
- 多维度分析：
    - 交叉（组合）维度分析，比如满足 条件 1+条件 2 的房子的套数、看房的次数、
    - 发布的时间 等等，注意分组 groupby 的使用，请同样完成分析+可视化+数据分析结果说明
    - 这个部分可以参考 Pandas 第 3 课的内容，更多角度的分析，有加分
    - 关联分析：
    - 我们关心房子的单价，有很多因素会影响价格，结合上面 2 个分析，看整体房价分布情况，再按照单维度和多维度(分组)看房价的状况，分析影响房价的核心因素是什么
    - 除掉分布分析，还可以看关联分析，可以查看 DataFrame 的 corr 函数，以及seaborn 当中的热力图
    
### 使用 sklearn/XGBoost/LightGBM 等工具完成房价预测模型建模
Tips：上部分可以为此处提供很好的分析数据依据，请牢记机器学习中的相关过程
- 数据预处理：
    - 包括异常值的检测(boxplot)与剔除、缺失值处理、数值型特征的处理方法、类别型特征的处理方法、组合特征生成、幅度缩放、特征选择等
- 机器学习建模：
    - 模型选择与网格搜索交叉验证，选择最优超参数
    - 模型效果评估：
    - 切分一部分数据为验证集，并在验证机上调超参与优化
    - 可以尝试多模型融合：
    - bagging 和 blending 等

## 数据清洗
- 拿到数据的时候，发现数据大量缺失，经过观察原来是爬虫爬取下来的数据只有部分解析了。需要自己进行解析处理。具体数据清洗过程参看[data_cleaning](data_cleaning.ipynb)

## 数据探索
- 数据探索是很重要的一步，决定了后续建模过程中如何进行特征工程，如何选择模型
- 主要分析手段有单维度分布分析、多维度分布分析、单/多维度与目标列的相关分析，同时在探索的过程中也会进行一些必要的数据处理。具体参看[EDA](task1/EDA.ipynb)

## 训练模型
- 主要步骤有缺失值处理、特征工程、模型选择和融合，以及预测。具体参看[Training&Predicting](task2/Training&Predicting.ipynb)

## 项目小结
  整个项目下来，对数据清洗，预处理，特征工程，模型选择和融合等一系列流程有了完整的了解，了解了不好数据处理技巧。
  特征工程是非常重要的一环，特征工程的好坏直接影响最终的模型学习效果。而特征工程需要基于扎实的业务知识
  模型融合是一个强悍的工程，就像“人多力量大”，集合集体的“智慧”，比单模型效果要好很多。
